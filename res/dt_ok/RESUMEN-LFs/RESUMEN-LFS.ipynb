```python
# Import packages to use
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams.update({'errorbar.capsize': 2})
```

```python
# Define the file to explore
file1='../RESUMEN-N100LF1.0/resumenLF1.csv'
file2='../RESUMEN-N100LF0.5/resumenLF0.5.csv'
# Import the data file
df1=pd.read_csv(file1, engine='python')
df05=pd.read_csv(file2, engine='python')
```

```python
# Get the theoretical nu value for each simulation
df1['nu-teo']=np.divide(-1.0,1.0+df1['beta'])
df05['nu-teo']=np.divide(-1.0,1.0+df05['beta'])
```

```python
fig,ax=plt.subplots(figsize=(12,6))

ax.scatter(df1['beta'],df1['nu'],marker='*',label=r'$\ell_f=1.0$')
ax.scatter(df05['beta'],df05['nu'],marker='x',label=r'$\ell_f=0.5$')

betaplot=np.linspace(min(df1['beta']),max(df1['beta']),1000)
nuplot=np.divide(-1.0,1.0+betaplot)
ax.plot(betaplot,nuplot,color='black',label='Theoretical data',linewidth=0.7,linestyle='--')

ax.axhline(-0.5,linestyle='-.',linewidth=0.7,color='red')

ax.set_xlabel(r'$\beta$',fontsize=30)
ax.set_ylabel(r'$\nu$',fontsize=30)

plt.grid(alpha=0.4)
plt.legend(fontsize=24)
plt.show()
plt.close()
```
