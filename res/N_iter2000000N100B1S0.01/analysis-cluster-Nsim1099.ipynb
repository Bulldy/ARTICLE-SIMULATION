```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
N_df=pd.read_csv('resumen-Nsim1099N_iter2000000N100B1S0.01.txt')
```

```python
N_df
```

```python
N_df.plot(x='log_t',y='logN')
```

```python
tail_df=N_df[(N_df['log_t']> 0) & (N_df['mean']>8)]
```

```python
tail_df
```

```python
tail_df.plot(x='log_t',y='logN')
```

```python
X=tail_df['log_t']
y=tail_df['logN']
X=sm.add_constant(X)
model=sm.OLS(y,X)
results=model.fit()
print(results.summary())
```
