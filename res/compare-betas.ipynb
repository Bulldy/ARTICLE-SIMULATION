```python
# Import packages to use
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
```

```python
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams.update({'errorbar.capsize': 2})
plt.rcParams.update({'font.size': 18})
```

```python
fileb1='resultados-simulaciones/resumen_N100_Niter10000000_beta1.0_dt1e-07_lf1.0_stf50_Nsimstot1044_mul10.csv'
fileb2='resultados-simulaciones/resumen_N100_Niter20000000_beta2.0_dt1e-07_lf1.0_stf50_Nsimstot1416_mul10.csv'
fileb4='resultados-simulaciones/resumen_N100_Niter25000000_beta4.0_dt4e-06_lf1.0_stf50_Nsimstot1000_mul10.csv'
```

```python
dfb1=pd.read_csv(fileb1)
dfb2=pd.read_csv(fileb2)
dfb4=pd.read_csv(fileb4)
```

```python
# Plot the entire log-data 
plt.figure(figsize=(12,6))
plt.rcParams.update({'font.size': 22})
plt.plot(dfb1.t,dfb1.N_avg,label=r'$\beta=1$')
plt.plot(dfb2.t,dfb2.N_avg,label=r'$\beta=2$')
plt.plot(dfb4.t,dfb4.N_avg,label=r'$\beta=4$')

plt.xlim(left=1e-4)
plt.ylim(bottom=10)
plt.yticks(np.arange(10,100,10))
plt.xlabel(r'$t$',fontsize=24)
plt.ylabel(r'$\langle N \rangle $',fontsize=24)
#plt.yaxis.set_major_formatter(mpl.ticker.ScalarFormatter())

plt.title(r'$N_0=%i \quad \ell_f=%g$'%(100,1),fontsize=26)
plt.grid(zorder=3)
plt.legend(fontsize=20)
plt.xscale('log')
plt.yscale('log')
plt.show()
plt.close()
```

```python
# Define the file to explore
file='resumenLF1.csv'
# Import the data file
df=pd.read_csv(file, engine='python')
```

```python
fig,ax=plt.subplots(figsize=(12,6))

ax.scatter(df['beta'],-df['nu'],label='Simulation data')
#ax.scatter(df['beta'],df['nu-teo'],label='Theoretical data')
ax.set_facecolor('white')
fig.set_facecolor('white')
betaplot=np.linspace(min(df['beta']),max(df['beta']),1000)
nuplot=np.divide(1.0,1.0+betaplot)
ax.plot(betaplot,nuplot,color='orange',label=r'$\nu\sim 1/(\beta+1)\qquad \beta \gg 1$')
ax.set_xlim(left=0,right=4)
ax.set_ylim(top=0.6,bottom=0.1)
ax.set_xlabel(r'$\beta$',fontsize=30)
ax.set_ylabel(r'$\nu$',fontsize=30)

plt.grid(alpha=0.4)
plt.legend(fontsize=24)
plt.show()
plt.close()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.scatter(df['beta'],-df['nu'],label='Simulation data')
ax.errorbar(x=df['beta'],y=-df['nu'],yerr=0.02,label=None,ls='none')
#ax.scatter(df['beta'],df['nu-teo'],label='Theoretical data')
ax.set_facecolor('white')
fig.set_facecolor('white')
betaplot=np.linspace(min(df['beta']),max(df['beta']),1000)
nuplot=np.divide(1.0,1.0+betaplot)
ax.plot(betaplot,nuplot,color='orange',label=r'$\nu\sim 1/(\beta+1)\qquad \beta \gg 1$')
ax.set_xlim(left=-0.1,right=4.1)
ax.set_ylim(top=0.8,bottom=0.1)
ax.set_xlabel(r'$\beta$',fontsize=30)
ax.set_ylabel(r'$\nu$',fontsize=30)

plt.grid(alpha=0.4)
plt.legend(fontsize=24)
plt.show()
plt.close()
```
